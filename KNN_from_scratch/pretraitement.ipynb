```python
import glob
import pandas as pd
import numpy as np

# get data file names
path =r'C:\Users\utilisateur\Desktop\Run_Questionnaire\DataSet'
filenames = glob.glob(path + "/DataSet*.csv")
# une boucle
dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
dataset_brut = pd.concat(dfs, ignore_index=True)
```

```python
#visualisations
print(dataset_brut)
```

```python
# verif le % de missing
for col in dataset_brut.columns:
    pct_missing = np.mean(dataset_brut[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# remplacer les vides par 0
dataset_brut = dataset_brut.fillna(0)
```

```python
#on uniformise majuscule et miniscule
dataset=dataset_brut.replace({'A':'a', 'B': 'b', 'C': 'c'})
dataset
```

```python
dataset['Q1']=dataset['Q1'].astype(str).astype(int)
```
